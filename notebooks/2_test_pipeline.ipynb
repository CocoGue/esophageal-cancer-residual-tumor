```python
from ct_residual_disease.pipelines.features_pipeline import (
    build_features_and_labels,
)

filename_csv_training = "../data/toy_dataset_training.csv"
X, y = build_features_and_labels(
    csv_path=str(filename_csv_training),
    feature_mode="combined",
    drop_first=True,
)

X.head(), y.value_counts()
```

```python
from ct_residual_disease.pipelines.training_pipeline import (
    train_logistic_pipeline,
)

results_training = train_logistic_pipeline(
    X,
    y,
    feature_mode="combined",
    scale_numerical=True,
    add_intercept=True,
    add_elasticnet_regularization=True,
    reglog_regularization=1.0,
)
```

```python
results_training["model"].get_coefficients()
results_training["model"].get_odds_ratios()
```

```python
from ct_residual_disease.pipelines.inference_pipeline import run_inference_pipeline

filename_csv_test = "../data/toy_dataset_test.csv"
inference_outputs = run_inference_pipeline(
    csv_path=filename_csv_test,
    model=results_training["model"],
    scaler=results_training["scaler"],
    threshold=results_training["threshold"],
    feature_mode="combined",
    encoding="logistic",
    drop_first=True,
)
```
